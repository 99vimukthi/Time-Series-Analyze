## KDE PLOT Production

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'AllDistrict_Production.csv'
data = pd.read_csv(file_path)


# Verify the column names to make sure 'Fertilizer' is the correct name
print(data.columns)

# Visualize the distribution of the 'Fertilizer' column
production_column = 'Production/kg'  # Replace with the correct column name if different

if production_column in data.columns:
    sns.kdeplot(data[production_column], fill=True)
    plt.title(f'KDE Plot of {production_column}')
    plt.xlabel(production_column)
    plt.ylabel('Density')
    plt.show()
else:
    print(f"Column '{production_column}' not found in the dataset.")
```

## Box plot for Production

```python
if production_column in data.columns:
    sns.boxplot(y=data[production_column])
    plt.title(f'Box Plot of {production_column}')
    plt.xlabel('production /kg')
    plt.ylabel('Values')
    plt.show()
else:
    print(f"Column '{production_column}' not found in the dataset.")
```
