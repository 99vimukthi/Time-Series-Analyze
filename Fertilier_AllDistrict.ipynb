## KDE Plot Fertilizer

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'AllDistrict_Production.csv'
data = pd.read_csv(file_path)


# Verify the column names to make sure 'Fertilizer' is the correct name
print(data.columns)

# Visualize the distribution of the 'Fertilizer' column
fertilizer_column = 'Fertilizer'  # Replace with the correct column name if different

if fertilizer_column in data.columns:
    sns.kdeplot(data[fertilizer_column], fill=True)
    plt.title(f'KDE Plot of {fertilizer_column}')
    plt.xlabel(fertilizer_column)
    plt.ylabel('Density')
    plt.show()
else:
    print(f"Column '{fertilizer_column}' not found in the dataset.")
```

## Box Plot Fertilizer

```python

if fertilizer_column in data.columns:
    sns.boxplot(y=data[fertilizer_column])
    plt.title(f'Box Plot of {fertilizer_column}')
    plt.xlabel('Fertilizer')
    plt.ylabel('Values')
    plt.show()
else:
    print(f"Column '{fertilizer_column}' not found in the dataset.")
```
