## KDE plot Rainy Days

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'AllDistrict_Production.csv'
data = pd.read_csv(file_path)


# Verify the column names to make sure 'Fertilizer' is the correct name
print(data.columns)

# Visualize the distribution of the 'Fertilizer' column
rainydays_column = 'Rainy days nos'  # Replace with the correct column name if different

if rainydays_column in data.columns:
    sns.kdeplot(data[rainydays_column], fill=True)
    plt.title(f'KDE Plot of {rainydays_column}')
    plt.xlabel(rainydays_column)
    plt.ylabel('Density')
    plt.show()
else:
    print(f"Column '{rainydays_column}' not found in the dataset.")
```

## Box Plot Rainy  Days No

```python
if rainydays_column in data.columns:
    sns.boxplot(y=data[rainydays_column])
    plt.title(f'Box Plot of {rainydays_column}')
    plt.xlabel('rainy days no')
    plt.ylabel('Values')
    plt.show()
else:
    print(f"Column '{rainydays_column}' not found in the dataset.")
```
